```python
groceries['st_lower'] = (groceries.HOUSE_NR_HI.astype('str')+ ' '+groceries.STREET+ ' '+groceries.STTYPE).apply(lambda x: x.lower())

groceries['st_lower_dir'] = (groceries.HOUSE_NR_HI.astype('str')+ ' '+ groceries.SDIR+' '+groceries.STREET+ ' '+groceries.STTYPE).apply(lambda x: x.lower())
```

```python
def dir_or_no_dir(df_1, df_1_street_cols, df_2, df_2_street_col, store_col, long, lat):
    st_array = []
    for i1,ss in df_1[df_1_street_cols].iterrows():
         for i2,s in df_2[[df_2_street_col, store_col,long,lat]].iterrows():
            if 'Av' in ss[0]:
                ss[0].replace('Av','Ave')
            if 'AV' in ss[0]:
                ss[0].replace('AV','AVE')
            if s[0].lower() == ss[0]:
                st_array.append([i1,s[0],ss[0],s[1],s[2],s[3]])
            if s[0].lower() == ss[1]:
                st_array.append([i1,s[0],ss[1],s[1],s[2],s[3]])
    return st_array
```

```python
wi_snap = pd.read_csv('WI.csv')
```

```python
st_array = dir_or_no_dir(groceries, ['st_lower_dir','st_lower'], wi_snap, 'ADDRESS', 'STORE_NAME','longitude','latitude')

```

```python
groceries_joined = groceries.reset_index().merge(pd.DataFrame(st_array, columns=['index','snap','groceries','store_name', 'longitude','latitude']), on='index')[['index','snap','groceries','store_name','st_lower_dir', 'longitude','latitude']]
```

```python
#"https://maps.googleapis.com/maps/api/geocode/json?address=" 
#address = "=1600+Amphitheatre+Parkway,+Mountain+View,+CA"
#"&key="
#g_api_key
def geo_locate(x):
    gl = "https://maps.googleapis.com/maps/api/geocode/json?address="
    address = x.replace(' ','+')+',+Milwaukee,+WI'
    response = req.get(gl+address+'&key='+g_api_key)
    res =response.json()
    result = res['results'][0]
    geodata = dict()
    geodata['lat'] = result['geometry']['location']['lat']
    geodata['lng'] = result['geometry']['location']['lng']
    geodata['address'] = result['formatted_address']
    print(geodata['address'])
    return [geodata['lng'], geodata['lat']]

```

```python
out = groc_tocsv.st_lower_dir_x.apply(lambda x: geo_locate(x))
```

```python
groc_geoloc =  pd.DataFrame(pd.DataFrame(out)['st_lower_dir_x'].tolist(), columns =['latG','longG'])
groc_tocsv['latG'] = groc_geoloc['latG']
groc_tocsv['longG'] = groc_geoloc['longG']
#pd.DataFrame(out)['st_lower_dir_x'].tolist()
```
